Zomato Data Analysis project


```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("Zomato data .csv")
```

```python
df
```

converting the data type


```python
def handleRate(value):
  value = str(value).split("/")
  value = value[0];
  return float(value)
df['rate']= df['rate'].apply(handleRate)
df.head()
```

```python
df.info()
```

Type of Restaurent

```python
df.head()
```

Q1. What type of restaurent do the majory of customer order from?

```python
sns.countplot(x=df['listed_in(type)'])
plt.title ("Type of Restaurent")

```

conclusion : Majorith of the restaurent falls in dinning category

Q2. How many votes has each type of restaurent received from customers?

```python
df.head()
```

```python
grouped_data =df.groupby('listed_in(type)')['votes'].sum()
result= pd.DataFrame({'votes':grouped_data})
plt.plot(result,c="green",marker='o')
plt.xlabel("Type of Restaurent ",c = "red" , size= 20)

plt.ylabel("votes",c ='red',size = 20)
```

Conclusion : Thi dinning restaurent has receive maximum votes

Q5. Which mode (online or offline) has received the maximum ratings?

```python
# which mode receives a maximum ratings
df.head()
```

Q6. Which type of restaurent recived more offline orders, so that Zomato can prove customers with some good offers?

```python
df.head()
```

```python
pivot_table =df.pivot_table (index= 'listed_in(type)',columns = 'online_order',aggfunc='size',fill_value=0)
sns.heatmap(pivot_table,annot=True ,cmap = "YlGnBu",fmt ='d')
plt.title("Heatmap")
plt.xlabel("Online_Order")
plt.ylabel("Listed In (Type)")
plt.show()
```

Conclusion : Dining restaurents primarily accept offline orders, whereas cafes primiarily receive online orders. This suggests that clients preffers orders in person at restaurents, but prefer online ordering at cafes.
